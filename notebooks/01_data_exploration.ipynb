```python
import riid
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Load the data we just generated
data_path = "../data/raw/train.h5"
ss = riid.SampleSet()
ss.load(data_path)

print(f"Loaded {ss.spectra.shape[0]} spectra with {ss.spectra.shape[1]} channels.")
print("Isotope classes present:", ss.get_labels().unique())

# Plot a few random samples
fig, ax = plt.subplots(figsize=(10, 6))
for i in range(3):
    idx = np.random.randint(0, len(ss.spectra))
    label = ss.get_labels().iloc[idx]
    ax.plot(ss.spectra.iloc[idx], label=f"Sample {idx} ({label})")

ax.set_title("Synthetic Gamma Spectra (Gross Counts)")
ax.set_xlabel("Channel (Energy)")
ax.set_ylabel("Normalized Intensity")
ax.legend()
plt.show()
```
